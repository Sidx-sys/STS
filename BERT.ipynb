```python
from sentence_transformers import SentenceTransformer
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader 
import numpy as np
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
import pickle

with open('data/train_data.pkl', 'rb') as f:
    train_data = pickle.load(f)
x_train = train_data['x']
y_train = train_data['y']

with open('data/test_data.pkl', 'rb') as f:
    test_data = pickle.load(f)
x_test = test_data['x']
y_test = test_data['y']
```

```python
x_train[0]
```

```python
formatted_train_data = [InputExample(texts=tr, label=(ts / 5)) for tr, ts in zip(x_train, y_train)]
```

```python
train_dataloader = DataLoader(formatted_train_data, shuffle=True, batch_size=32)
train_loss = losses.CosineSimilarityLoss(model)
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=1, warmup_steps=100)
```

```python
def get_embeddings(sent_1, sent_2):
    embeddings1 = model.encode(sent_1, convert_to_tensor=True).cpu().numpy()
    embeddings2 = model.encode(sent_2, convert_to_tensor=True).cpu().numpy()

    return embeddings1, embeddings2
```

```python
with open('data/test_data.pkl', 'rb') as f:
    test_data = pickle.load(f)
x_test = test_data['x']
y_test = test_data['y']
```

```python
def cosine_similarity(vec_1, vec_2):
    return vec_1@vec_2.T/(np.linalg.norm(vec_1) * np.linalg.norm(vec_2))
```

```python
preds = []
for x in x_test:
    vec_1, vec_2 = get_embeddings(x[0], x[1])
    preds.append(5 * cosine_similarity(vec_1, vec_2))
```

```python
import scipy
pearson_score, _ = scipy.stats.pearsonr(preds, y_test)
pearson_score
```
